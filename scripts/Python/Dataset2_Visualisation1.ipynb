```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import seaborn as sns
import os
```

```python
df = pd.read_csv("../../clean_data/Dataset2/BRYANT.csv")
df1 = pd.read_csv("../../clean_data/Dataset2/HAMILTON.csv")
df2 = pd.read_csv("../../clean_data/Dataset2/HIGH.csv")
```

```python
# Function to create directory if it doesn't exist
def ensure_dir(file_path):
    directory = os.path.dirname(file_path)
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
# Convert 'Date_Time' to datetime and extract the hour
df['Date_Time'] = pd.to_datetime(df['Date_Time'])
df['Hour of Day'] = df['Date_Time'].dt.hour

# Line Graph
line_data = df.groupby('Hour of Day')['Energy'].mean()
plt.figure(figsize=(12, 6))
plt.plot(line_data.index, line_data.values,color='lightsalmon', marker='o')
plt.title('Bryant Station - Average Energy Usage by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Energy Usage (kWh)')
plt.grid(True)
plt.xticks(range(24))
plot_path = '../../visualisations/Average_Energy_Usage_by_Hour/Bryant_Station.png'
ensure_dir(plot_path)
plt.savefig(plot_path)
plt.show()

# Heatmap
df_filtered = df[df['Day_of_Week'] != 'T']
days = ["Sunday", "Saturday", "Friday", "Thursday", "Wednesday", "Tuesday", "Monday"]
df_filtered['Day_of_Week'] = pd.Categorical(df_filtered['Day_of_Week'], categories=days, ordered=True)
heatmap_data = df_filtered.pivot_table(values='Energy', index='Day_of_Week', columns='Hour of Day', aggfunc='mean')
plt.figure(figsize=(14, 7))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, cbar_kws={'label': 'Energy Usage (kWh)'})
plt.title('Bryant Station - Heatmap of Average Energy Usage by Time and Day of Week')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plot_path = '../../visualisations/Average_Energy_Usage_by_Time_and_Day_of_Week/Bryant_Station.png'
ensure_dir(plot_path)
plt.savefig(plot_path)
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Hour of Day'], df['Energy'], color='lightsalmon')
plt.title('Bryant Station - Scatter Plot of Energy Usage by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Energy Usage (kWh)')
plt.xticks(range(24))
plt.grid(True)
plot_path = '../../visualisations/Energy_Usage_by_Hour/Bryant_Station.png'
ensure_dir(plot_path)
plt.savefig(plot_path)
plt.show()
```
