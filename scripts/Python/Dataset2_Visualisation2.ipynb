```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import seaborn as sns
import os
```

```python
df = pd.read_csv("../../clean_data/Dataset2/BRYANT.csv")
df1 = pd.read_csv("../../clean_data/Dataset2/HAMILTON.csv")
df2 = pd.read_csv("../../clean_data/Dataset2/HIGH.csv")
```

```python
# Function to create directory if it doesn't exist
def ensure_dir(file_path):
    directory = os.path.dirname(file_path)
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
df['Date_Time'] = pd.to_datetime(df['Date_Time'])

# Resampling and Rolling Mean
# Resample the data to a weekly frequency
weekly_energy = df.resample('W', on='Date_Time')['Energy'].sum()

# Apply a rolling mean with a window of 4 weeks for smoothing
rolling_energy = weekly_energy.rolling(window=4).mean()

# Plotting the smoothed data
plt.figure(figsize=(15, 8))
plt.plot(rolling_energy.index, rolling_energy.values,color="lightsalmon", marker='o', linestyle='-')
plt.title('Bryant Station - Total Energy Demand Over Time (Weekly, Smoothed)')
plt.xlabel('Date')
plt.ylabel('Total Energy (kWh)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plot_path = '../../visualisations/Total_Energy_Demand_Over_Time/Bryant_Station.png'
ensure_dir(plot_path)
plt.savefig(plot_path)
plt.show()

# Cumulative Sum
# Calculate the cumulative sum of the energy
cumulative_energy = weekly_energy.cumsum()

# Plotting Cumulative Line Graph
plt.figure(figsize=(15, 8))
plt.plot(cumulative_energy.index, cumulative_energy.values,color="lightsalmon",marker='o', linestyle='-')
plt.title('Bryant Station - Cumulative Energy Demand Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Energy (kWh)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plot_path = '../../visualisations/Cumulative_Energy_Demand_Over_Time/Bryant_Station.png'
ensure_dir(plot_path)
plt.savefig(plot_path)
plt.show()
```
