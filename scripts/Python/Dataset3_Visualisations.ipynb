```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

df = pd.read_csv("../../clean_data/Dataset3/BRYANT.csv")
df1 = pd.read_csv('../../clean_data/Dataset3/HAMILTON.csv')
df2 = pd.read_csv('../../clean_data/Dataset3/HIGH.csv')
```

```python
# Function to create directory if it doesn't exist
def ensure_dir(file_path):
    directory = os.path.dirname(file_path)
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
frequency_distribution = df['Free_Station_Count'].value_counts()

# Get the range of Free_Station_Counts
free_station_count_range = range(frequency_distribution.index.min(), frequency_distribution.index.max() + 1)

extended_range = range(0, frequency_distribution.index.max() + 2)

# Plotting the frequency distribution as a bar chart
plt.figure(figsize=(10, 6))
plt.bar(extended_range[:-1], frequency_distribution.reindex(extended_range[:-1], fill_value=0),
        color='lightsalmon', edgecolor='black')

# Setting the x-axis ticks to include 0 and extend to the max value
plt.xticks(extended_range[:-1])

plt.title('Bryant Station - Frequency Distribution of Free Station Count')
plt.xlabel('Free Station Count')
plt.ylabel('Frequency (amount of hours)')
plt.grid(axis='y', alpha=0.75)
plot_path = '../../visualisations/Frequency_Distribution_of_Free_Station_Count/Bryant_Station.png'
ensure_dir(plot_path)
plt.savefig(plot_path)

# Show plot
plt.show()

df['Date_Time'] = pd.to_datetime(df['Date_Time'])

# Group the data by hour and calculate the average
df['hour'] = df['Date_Time'].dt.hour
average_per_hour = df.groupby('hour')['Free_Station_Count'].mean()

# Plotting the average number of stations available for each hour
plt.figure(figsize=(10, 6))
average_per_hour.plot(kind='bar', color='lightsalmon', edgecolor='black')

plt.title('Bryant Station - Average Number of Stations Available for Each Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Number of Free Stations')
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=0)

# Show plot
plot_path = '../../visualisations/Average_Number_of_Stations_Available_for_Each_Hour/Bryant_Station.png'
ensure_dir(plot_path)
plt.savefig(plot_path)
plt.show()
```
